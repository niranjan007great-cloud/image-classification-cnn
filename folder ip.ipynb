```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Specify the directory containing the images
image_dir = '/content/drive/MyDrive/Mainipfolder/Train/Samsung' # Replace with your folder path

# List all files in the directory
image_files = [f for f in os.listdir(image_dir) if os.path.isfile(os.path.join(image_dir, f))]

# Display all images
print(f"Displaying all {len(image_files)} images from {image_dir}:")

for i, image_file in enumerate(image_files):
    image_path = os.path.join(image_dir, image_file)
    try:
        img = mpimg.imread(image_path)
        plt.imshow(img)
        plt.title(image_file)
        plt.axis('off') # Hide axes
        plt.show()
    except Exception as e:
        print(f"Could not read image {image_file}: {e}")

if not image_files:
    print(f"No files found in the directory: {image_dir}")
```
